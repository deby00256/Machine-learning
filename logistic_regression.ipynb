```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
df=pd.read_csv("Social_Network_Ads.csv")
x=df.iloc[:,[2,3]]
y=df.iloc[:,[4]]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
model=LogisticRegression()
model.fit(x_train,y_train)

```

```python
#normalization
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(634)
x1=np.random.randint(18,60,30)
x1.sort()
x2=np.random.randint(15000,150000,30)
plt.plot(x1,x2)
```

```python
x1min=min(x1)
x1max=max(x1)
x2min=min(x2)
x2max=max(x2)
x1norm=(x1-x1min)/(x1max-x1min)
x2norm=(x2-x2min)/(x2max-x2min)
plt.plot(x1norm,x2norm)
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
x_train=scale.fit_transform(x_train)
x_tes=scale.fit_transform(x_test)
yped=model.predict(x_tes)
yped
from sklearn.metrics import accuracy_score
accuracy_score(y_test,yped)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,yped))
```
